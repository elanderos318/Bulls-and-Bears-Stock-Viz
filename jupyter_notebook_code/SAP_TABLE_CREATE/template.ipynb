```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine, func
```

```python
from sqlalchemy.ext.automap import automap_base
```

```python
from flask import Flask, jsonify, render_template, request, redirect, url_for, session
```

```python
from sqlalchemy.orm import Session
```

```python
engine = create_engine('sqlite:///../../db/bulls_bears.sqlite', echo=False)
```

```python
# Declare a Base using automap_base()
Base = automap_base()
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
SP500 = Base.classes.sp_data_20200320
```

```python
engine.execute("SELECT * FROM sp_data_20200320 LIMIT 10").fetchall()
```

```python
conn = engine.connect()
```

```python
pd.read_sql("SELECT * FROM sp_data_20200320 LIMIT 10", con = conn).reset_index(drop = True)
```

```python
session = Session(engine)
```

```python
session.query(SP500.date, SP500.open, SP500.high, SP500.low, SP500.close, SP500.adj_close, SP500.volume).limit(5).all()
```

```python
home = session.query(SP500.date).all()
```

```python
for query in home:
    print(query)
```
